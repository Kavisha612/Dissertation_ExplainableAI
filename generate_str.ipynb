```python
import json
from pathlib import Path

def print_repo(root=".", indent=0, skip={"__pycache__", ".git", ".png" ".ipynb_checkpoints", ".venv", "venv", "env", "node_modules"}):
    for path in sorted(Path(root).iterdir()):
        if path.name in skip:
            continue
        prefix = "  " * indent + ("📁 " if path.is_dir() else "📄 ")
        print(f"{prefix}{path.name}")
        if path.is_dir():
            print_repo(path, indent + 1, skip)
        elif path.suffix == ".ipynb":
            nb = json.load(open(path, encoding="utf-8"))
            for cell in nb["cells"]:
                if cell["cell_type"] == "markdown" and cell["source"]:
                    line = cell["source"][0].strip()
                    if line.startswith("#"):
                        print("  " * (indent + 1) + "↳ " + line.lstrip("#").strip())

print_repo("")  # adjust to point at your repo root
```
